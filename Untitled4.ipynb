```python
# 🔷 Mental Health Prediction ML Project (Single Code - Google Colab)

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Upload dataset
from google.colab import files
uploaded = files.upload()

# Load dataset (example: mental_health.csv)
data = pd.read_csv("mental_health.csv")

# Show dataset
print("First 5 rows:")
print(data.head())

# 🔷 Data Preprocessing

# Convert categorical data to numbers
le = LabelEncoder()

for column in data.columns:
    if data[column].dtype == 'object':
        data[column] = le.fit_transform(data[column])

# Check for missing values
print("\nMissing Values:\n", data.isnull().sum())

# Fill missing values (simple method)
data = data.fillna(data.mean())

# 🔷 Features and Target
# Assume 'treatment' is target (0 = No, 1 = Yes)
X = data.drop("treatment", axis=1)
y = data["treatment"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# 🔷 Model Training
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# 🔷 Prediction
y_pred = model.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("\nModel Accuracy:", accuracy)

# 🔷 Custom Prediction (Example)
print("\nCustom Prediction:")
sample = [X.iloc[0].values]   # using first row as sample
prediction = model.predict(sample)

if prediction[0] == 1:
    print("Needs Mental Health Treatment")
else:
    print("No Treatment Needed")

# 🔷 Visualization
plt.hist(y)
plt.title("Mental Health Treatment Distribution")
plt.xlabel("Treatment (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```
